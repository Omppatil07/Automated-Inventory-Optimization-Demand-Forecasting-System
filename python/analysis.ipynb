```python
df = df.merge(inventory, on = 'product_ID', how = 'left')
df
```

```python
print(df)
```

```python
# Rename and Drop
```

```python
df = df.rename(columns={"selling_price_x": "selling_price"})
df = df.drop(columns=["selling_price_y"])
```

```python
df
```

```python
df = df.rename(columns={"Lead_Time": "lead_time"})
df
```

```python
# Feature Engineering
```

```python
# Total sales
```

```python
df["Total_Sales_Product"] = df.groupby("product_ID")["units_sold"].transform("sum")
```

```python
df = df.rename(columns={"Total_Sales_Product": "total_sales_product"})
```

```python
df
```

```python
# Daily Average Sales
```

```python
df['daily_avg_sales'] = df.groupby('product_ID')['units_sold'].transform('mean').round(2)
df
```

```python
# Days of Stock
```

```python
df['days_of_stock'] = (df['stock_available'] / df['daily_avg_sales']).round()
df
```

```python
# Inventory Turnover
```

```python
df["inventory_turnover"] = (df["total_sales_product"] / df["stock_available"]).round()
```

```python
df
```

```python
# Risk Classification
```

```python
def risk_category(days):
    if days < 15:
        return "High Risk"
    elif days < 25:
        return "Medium Risk"
    else:
        return "Low Risk"

df["inventory_risk"] = df["days_of_stock"].apply(risk_category)
```

```python
df
```

```python
(df[df["inventory_risk"] == "Medium Risk"]).count()
```

```python
# Reorder Point Calculation
```

```python
df = df.drop(columns = ['reorder_point'])
```

```python
df
```

```python
safety_stock = 30

df["reorder_point"] = ((df["daily_avg_sales"] * df["lead_time"]) + safety_stock).round()
```

```python
df
```

```python
# Reorder Decision (Automation Logic)
```

```python
df["reorder_flag"] = df.apply(
    lambda x: "Reorder Needed" if x["stock_available"] < x["reorder_point"] else "Sufficient",
    axis=1
)
```

```python
df
```

```python
df.to_csv("final_inventory_analysis.csv", index=False)
```
